```python
df1 = pd.concat([df_1988, df_1989, df_1990, df_2009, df_2014, df_2015,
                df_2016, df_2017, df_2018, df_2019, df_2020,
                df_2021], axis=0)
df1 = df1.reset_index()
df1 = df1.drop(['index','#','Unnamed: 4','Unnamed: 6','Unnamed: 3',
                'Unnamed: 5'], axis=1)
df1.head(10)
```

```python
print("The dataset until now has {} rows and {} cols.".format(df1.shape[0],
                                                              df1.shape[1]))
```

```python
df1 = df1.replace(to_replace="República Dominicana", value="Rep. Dominicana")
df1 = df1.replace(to_replace="Combinado Jugadores Estadounidenses", 
                  value="Estados Unidos")
df1.head(100)
```

```python
df1 = df1.rename(columns={"Competición": "Competition",
                        "Fecha": "Date",
                        "Lugar": "Place",
                        "Resultado": "Score",
                        "Visitante": "Visitor"})
df1.head(10)
```

```python
df1["L_Score"] = df1['Score'].astype(str)
df1["V_Score"] = df1['Score'].astype(str)

df1["L_Score"] = df1["Score"].str.split(':', 1).str[0].str.strip()
df1["V_Score"] = df1['Score'].str.split(':', 1).str[1].str.strip()
```

```python
# cols = list(df1.columns.values)
# cols
df1 = df1[['Date','Place','Local','L_Score','V_Score','Visitor','Competition']]
df1.head(15)
```

```python
# The data from these years has different score format, we'll do another df
df2 = pd.concat([df_1991, df_1992, df_1997, df_2001, df_2002,
                 df_2003, df_2004, df_2005, df_2006, df_2007,
                 df_2008, df_2011, df_2012, df_2013], axis=0)
df2 = df2.reset_index()

# Let's rename this column and then merge with the rest of the dataframe
df_2010 = df_2010.rename(columns={"Competición'": "Competición"})

df2 = pd.concat([df2, df_2010], axis=0)
df2 = df2.reset_index(drop=True)
df2 = df2.drop(['index', '#'], axis=1)
```

```python
df2 = df2.rename(columns={"Competición": "Competition",
                        "Fecha": "Date",
                        "Lugar": "Place",
                        "Resultado": "L_Score",
                        "Resultado.1": "V_Score",
                        "Visitante": "Visitor"})
df2.head(10)
```

```python
df2 = df2.replace(to_replace="Islas Vírgenes Estadounidenses", 
                  value="Islas Vírgenes")
```

```python
# Let's join everything and see the dimensions
df = pd.concat([df1, df2], axis=0)
print("The dataset until now has {} rows and {} cols.".format(df.shape[0],
                                                              df.shape[1]))
```

```python
# We'll create columns to record Venezuela's team wins and losses
df["Win"] = 0
df["Loss"] = 0

results = np.where(
    ((df['Visitor'] == 'Venezuela') & (df['V_Score'] > df['L_Score'])) |
    ((df['Local'] == 'Venezuela') & (df['L_Score'] > df['V_Score'])),
    'win', 'loss')

# We can quickly check the team's record (count of total wins and losses)
record = {'wins': np.sum(results == 'win'), 'losses': np.sum(results == 'loss')}
record
```

```python
df["Local"] = df["Local"].str.strip()
df["Visitor"] = df["Visitor"].str.strip()

df["V_Score"] = pd.to_numeric(df["V_Score"], downcast="integer")
df["L_Score"] = pd.to_numeric(df["L_Score"], downcast="integer")

# Setting 1 for wins and 0 for losses according to the resulting score 
def get_results(df):
  for row in df:
    if ((df['Visitor'] == 'Venezuela') & (df['V_Score'] > df['L_Score'])):
      df["Win"] = 1
    elif ((df['Local'] == 'Venezuela') & (df['L_Score'] > df['V_Score'])):
      df["Win"] = 1
    else: 
      df["Loss"] = 1
  return df

df = df.apply(get_results, axis='columns')
```

```python
df = df.reset_index(drop=True)
df.head(10)
```

```python
# Formatting date
df['Date'] = df['Date'].str.replace(' de ','')

# Months
df['Date'] = df['Date'].str.replace('enero','01')
df['Date'] = df['Date'].str.replace('febrero','02')
df['Date'] = df['Date'].str.replace('marzo','03')
df['Date'] = df['Date'].str.replace('abril','04')
df['Date'] = df['Date'].str.replace('mayo','05')
df['Date'] = df['Date'].str.replace('junio','06')
df['Date'] = df['Date'].str.replace('julio','07')
df['Date'] = df['Date'].str.replace('agosto','08')
df['Date'] = df['Date'].str.replace('septiembre','09')
df['Date'] = df['Date'].str.replace('octubre','10')
df['Date'] = df['Date'].str.replace('noviembre','11')
df['Date'] = df['Date'].str.replace('diciembre','12')

# Days
def format_day(df):
  for row in df:
    if len(df["Date"]) == 7:
      df["Date"] = '0' + df["Date"]
  return df

df = df.apply(format_day, axis='columns')
df
```

```python
# There are only two rows with missing data (day of the game), index 128
# and 237, let's drop them since I could not find the info about those games..
df = df.drop([df.index[128],df.index[237]])
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d%m%Y', utc=False)
df['Date'] = df['Date'].dt.date
df
```

```python
df = df.sort_values(by="Date", axis=0, ascending=True)
df = df.reset_index(drop=True)
df
```
